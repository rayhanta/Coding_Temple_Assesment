```python
## Data Cleaning
```

```python
# Remove any columns not necessary for the analysis questions below.
playlist_1_cleaned = playlist_1[["album", "artists", "duration_ms", "explicit", "name", "popularity", "preview_url"]]
playlist_2_cleaned = playlist_2[["album", "artists", "duration_ms", "explicit", "name", "popularity", "preview_url"]]
```

```python
# Change datatypees, rename, and reorder columns as you see fit.
playlist_1_cleaned['duration'] = pd.to_datetime(playlist_1_cleaned['duration_ms'], unit = 'ms')
playlist_2_cleaned['duration'] = pd.to_datetime(playlist_2_cleaned['duration_ms'], unit = 'ms')
```

```python
# Transform the duration_ms column from milliseconds to minutes:seconds
playlist_1_cleaned['duration'] = playlist_1_cleaned['duration'].dt.time
playlist_1_cleaned.dtypes

playlist_2_cleaned['duration'] = playlist_2_cleaned['duration'].dt.time
playlist_2_cleaned.dtypes
```

```python
# Should the artists column be transformed? Explain your choice and follow through.

# I believe that the artist column should not be transformed, as we are assuming the song has equal contributions
# from all artists involved. (Think how a band would not be able to perform fully without a member like
# a guitarist or a bassist.)
```

```python
## Extraction to SQL database
```

```python
# Label which playlist each song belongs to
playlist_1_cleaned['playlist'] = 1
playlist_2_cleaned['playlist'] = 2
```

```python
# Drop duplicates
playlist_1_cleaned_2 = playlist_1_cleaned.drop_duplicates()
playlist_2_cleaned_2 = playlist_2_cleaned.drop_duplicates()
```

```python
# Drop null values
playlist_1_cleaned_2 = playlist_1_cleaned_2.dropna()
playlist_2_cleaned_2 = playlist_2_cleaned_2.dropna()
```

```python
# Combine these 2 dataframes into 1
playlist_combined = pd.concat([playlist_1_cleaned_2, playlist_2_cleaned_2])
playlist_combined = playlist_combined.reindex(["preview_url","album", "artists", "duration", "explicit", "name", "popularity"], axis = 1)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Create a visualization of song runtime.
playlist_combined_anv_2 = pd.concat([playlist_1_cleaned_2, playlist_2_cleaned_2])
top_5_runtime = playlist_combined_anv_2.sort_values(by = ['duration_ms'], ascending = False)[:5]
top_5_runtime.plot(kind = 'bar', x = 'name', y = 'duration_ms', title = 'runtime')
```

```python
# What is the average runtime of songs in each playlist?
playlist_combined_anv.groupby('playlist').mean()['duration_ms']/1000

# 218.44 seconds in playlist 1
# 217.65 seconds in playlist 2
```

```python
# Which playlist has a longer total runtime?
playlist_combined_anv.groupby('playlist').sum()['duration_ms']/1000

# Playlist 1 has a longer total runtime
```

```python
# Which playlist contains more popular music?
playlist_combined_anv.groupby('playlist').sum()['popularity']

# based on total amount of likes
# playlist 2 is more popular.
```

```python
# Create a visualization of song popularity.
top_5_popular = playlist_combined_anv_2.sort_values(by = ['popularity'], ascending = False)[:5]
top_5_popular.plot(kind = 'bar', x = 'name', y = 'popularity', title = 'popularity')
```

```python
# Which playlist contains more duplicate songs?
playlist_1_cleaned_no_duplicates = playlist_1_cleaned.drop_duplicates()
playlist_2_cleaned_no_duplicates = playlist_2_cleaned.drop_duplicates()

print("playlist 1 dupes: ",len(playlist_1_cleaned) - len(playlist_1_cleaned_no_duplicates))
print("playlist 2 dupes: ",len(playlist_2_cleaned) - len(playlist_2_cleaned_no_duplicates))

# Playlist 2 contains more duplicate songs
```

```python
# Which song is present the most times across both playlists?
playlist_combined_anv.name.mode()

# Drift and Grey Luh is present the most times across both playlists
```

```python
# Which artist has the most songs in each playlist?
print(playlist_1_cleaned_no_duplicates.artists.mode())
print(playlist_2_cleaned_no_duplicates.artists.mode())

# Playlist 1: Quinn XCII
# Playlist 2: 'Trettmann', 'KitschKrieg', 'SFR'
```

```python
# Create a visualization of songs per artist for each playlist.
top_5_artist = playlist_combined_anv_2.groupby('artists').count()
top_5_artist = top_5_artist.sort_values(by = ['duration_ms'], ascending = False)[:5]

top_5_artist.plot(kind = 'bar', y = 'album', title = "number of songs per artist for each playlist")
```

```python
# Which artist has the most songs from different albums across both playlists?
playlist_grouped_by_artist = playlist_combined_anv.groupby('artists').count()
sorted_playlist = playlist_grouped_by_artist.sort_values('duration', ascending = False)
print(sorted_playlist)

# Quinn XCII has the most songs from different albums across both playlists.
```
